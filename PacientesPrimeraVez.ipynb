```python
sustancias = ['TABACO', 'ALCOHOL', 'CANNABIS/MARIGUANA',
       'COCAINA', 'ESTIMULANTES/ANFETAMINAS',
       'ESTIMULANTES DE TIPO ANFETAMINICO', 'ESTIMULANTES OTROS', 'INHALABLES',
       'ALUCINOGENOS', 'DEPRESORES DEL SNC', 'OPIO', 'OPIOIDES',
       'OPIO/FENTANILO', 'SUST CON UTIL MEDICA', 'OTROS', 'NO CLASIFICADAS',
       'SIN CONSUMO',]

for sustancia in sustancias:
    sns.countplot(data=df_limpio, x=sustancia, hue=sustancia, palette='rocket', legend=False)
    plt.title(f'Consumo de {sustancia}')
    plt.show()
```

```python
columnas_categoricas = ['SEXO', 'EDAD', 'ESTADO_CONYUGAL', 'ESCOLARIDAD','OCUPACION', 'TIPO_CONSUMO']

# Número de gráficos por fila y columna
n_filas = 1  # 1 fila
n_columnas = 2  # 2 columnas

# Iterar sobre las columnas categóricas en grupos de 2
for i in range(0, len(columnas_categoricas), n_filas * n_columnas):
  
    fig, axes = plt.subplots(n_filas, n_columnas, figsize=(15, 6))
    fig.suptitle(f'Distribución por MOTIVO', fontsize=16)
    
  
    axes = axes.flatten()
    
    # Graficar cada columna categórica en un subgráfico
    for j, columna in enumerate(columnas_categoricas[i:i + n_filas * n_columnas]):
        sns.countplot(data=df_limpio, x=columna, hue=df_limpio['MOTIVO'], palette='rocket', ax=axes[j])
        axes[j].set_title(f'{columna} por MOTIVO')
        axes[j].legend(title='MOTIVO', bbox_to_anchor=(1.05, 1), loc='upper left')
        axes[j].tick_params(axis='x', rotation=45)  # Rotar etiquetas si es necesario
    
    # Ajustar el espaciado entre subgráficos
    plt.tight_layout()
    plt.show()
```

```python
# Matriz de correlación con respecto a las columnas de sustancia:

matriz_correlacion = df_limpio[sustancias].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacion, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5)
plt.title('Matriz de Correlación entre Sustancias')
plt.show()
```
