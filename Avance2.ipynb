Distribución del total suicidios registrados por año 

```python
# Extraer 'Total' para todos los años
df_total = df.xs('Total', axis=1, level=1)

df_total = pd.concat([df.iloc[:, :2], df_total], axis=1)

df_total[df_total['Grupos quinquenales de edad', 'Grupos quinquenales de edad'] == 'Total']
```

```python
df_total.iloc[0]
```

```python
df_total = df_total.apply(pd.to_numeric, errors='coerce')
```

```python


df_total.iloc[0,2:].plot(kind='bar', figsize=(8, 4))
plt.title("Distribución de registros totales por año")
plt.xlabel("Años")
plt.ylabel("Cantidad de registros")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Registros de hombres 

```python
df_total_H = df.xs('Hombres', axis=1, level=1)
df_total_H = pd.concat([df.iloc[:, :2], df_total_H], axis=1)
df_total_H[df_total_H['Grupos quinquenales de edad', 'Grupos quinquenales de edad'] == 'Total'].head()
```

Registros de Mujeres 

```python
df_total_M = df.xs('Mujeres', axis=1, level=1)
df_total_M = pd.concat([df.iloc[:, :2], df_total_M], axis=1)
df_total_M[df_total_M['Grupos quinquenales de edad', 'Grupos quinquenales de edad'] == 'Total'].head()
```

```python


valores_hombres = df_total_H.iloc[0, 2:]
valores_mujeres = df_total_M.iloc[0, 2:]

anios = valores_hombres.index  
x = np.arange(len(anios))      
width = 0.4                

plt.figure(figsize=(10, 5))

# Barras para Hombres (desplazadas a la izquierda)
plt.bar(x - width/2, valores_hombres, width, label='Hombres')

# Barras para Mujeres (desplazadas a la derecha)
plt.bar(x + width/2, valores_mujeres, width, label='Mujeres')

plt.xticks(x, anios, rotation=45)
plt.xlabel("Años")
plt.ylabel("Registros Totales")
plt.title("Comparativa de registros (Hombres vs. Mujeres)")
plt.legend()
plt.tight_layout()
plt.show()
```
